```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("Библиотеки импортированы.")

filename = 'final_sensor_table_sorted.csv'
try:
    df = pd.read_csv(filename, index_col='timestamp', parse_dates=True)
    print(f"Файл '{filename}' загружен.\n")
except FileNotFoundError:
    print(f"Файл '{filename}' не найден.")
    raise
#корреляции  
print('Расчет корреляций:\n')
absolute_correlation_matrix = df.corr(method='pearson').abs()
power_column = 'Vessel Propeller Shaft Mechanical Power'
power_correlations = absolute_correlation_matrix[power_column].sort_values(ascending=False)
print(power_correlations[1:11].round(3))
top_power_features = set(power_correlations[1:11].index)
print('\n')
speed_column = 'Vessel Hull Through Water Longitudinal Speed'
speed_correlations = absolute_correlation_matrix[speed_column].sort_values(ascending=False)
print(speed_correlations[1:11].round(3))
top_speed_features = set(speed_correlations[1:11].index)

# пересечения
common_features = top_power_features.intersection(top_speed_features)

if not common_features:
    print("\nНет общих признаков.")
else:
    common_feature_scores = {}
    for feature in common_features:
        score = power_correlations[feature] + speed_correlations[feature]
        common_feature_scores[feature] = score
    sorted_common_features = sorted(common_feature_scores, key=common_feature_scores.get, reverse=True)
    print('\n')
    
    print(f"Найдено {len(sorted_common_features)} общих признаков, отсортированных по суммарной силе связи:")
    for i, feature in enumerate(sorted_common_features):
        print(f"  {i+1}: {feature} (Суммарная корреляция: {common_feature_scores[feature]:.3f})")
        
print("\nКонец!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

filename = 'final_sensor_table_sorted.csv'
try:
    df = pd.read_csv(filename)
    df['timestamp'] = pd.to_datetime(df['timestamp'], format='ISO8601')
    df.set_index('timestamp', inplace=True)
    print(f"Файл загружен. Для анализа доступно {len(df)} строк.")
except FileNotFoundError:
    print(f"Файл '{filename}' не найден.")
    raise
    
features_to_segment_by = [
    'Vessel Propeller Shaft Torque',
    'Main Engine Rotational Speed',
    'Main Engine Turbocharger Rotational Speed',
    'Vessel Hull Over Ground Speed',
    'Vessel Propeller Shaft Rotational Speed',
    'Main Engine Fuel Load %',
    'Main Engine Fuel Oil Inlet Mass Flow',
    'Main Engine Scavenging Air Pressure',
    'Vessel Propeller Shaft Thrust Force'
]

features_to_analyze_in_summary = [
    'Vessel Propeller Shaft Mechanical Power',
    'Vessel Hull Through Water Longitudinal Speed' 
]

x_axis_feature = 'Vessel Propeller Shaft Mechanical Power'
y_axis_feature = 'Vessel Hull Through Water Longitudinal Speed'

for main_feature in features_to_segment_by:
    print(f"Анализ по перцентилям: '{main_feature}'")
    
    analysis_df = df.copy()
    key_cols = [main_feature, x_axis_feature, y_axis_feature] + features_to_analyze_in_summary
    analysis_df.dropna(subset=list(set(key_cols)), inplace=True)

    num_groups = 10
    labels = [f"{i*10}-{(i+1)*10}-й перцентиль" for i in range(num_groups)]

    analysis_df['rank'] = analysis_df[main_feature].rank(pct=True)
    analysis_df['group'] = pd.cut(analysis_df['rank'], bins=num_groups, labels=labels, include_lowest=True)
    
    print(f"\nРазделили {len(analysis_df)} строк на {num_groups} групп по '{main_feature}'.")

    # границы
    print()
    decile_points = np.linspace(0, 1, num_groups + 1)
    decile_boundaries = analysis_df[main_feature].quantile(decile_points)
    for i in range(num_groups):
        lower_bound = decile_boundaries.iloc[i]
        upper_bound = decile_boundaries.iloc[i+1]
        print(f"  - {labels[i]}: от {lower_bound:.2f} до {upper_bound:.2f}")

    # сводная таблица
    percentile_summary = analysis_df.groupby('group', observed=False)[features_to_analyze_in_summary].agg(['mean', 'size'])
    print("\nСредние показатели для каждой группы:")
    print(percentile_summary)
    
    # график
    if len(analysis_df) > 75000:
        plot_df = analysis_df.sample(75000, random_state=42)
    else:
        plot_df = analysis_df
    
    plt.figure(figsize=(14, 8))
    sns.scatterplot(data=plot_df,
                    x=x_axis_feature,
                    y=y_axis_feature,
                    hue='group', 
                    hue_order=labels,
                    palette='viridis',
                    alpha=0.6,
                    s=25)
    
    plt.title(f'Зависимость скорость - мощность')
    plt.xlabel('Vessel Propeller Shaft Mechanical Power')
    plt.ylabel('Vessel Hull Through Water Longitudinal Speed')
    plt.grid(True)
    plt.legend(title=f'Перцентиль по "{main_feature}"', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

print("\nКонец!")
```
